# Creazione dataset Predictions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
raw_data_folder = Path(r"C:\Users\Emanuele\Documents\Progetti Python\combining_strategies_ml\data\offline")
```

```python
df_complete = pd.DataFrame()

# Loop through all files in the folder
for file in raw_data_folder.iterdir():
    # Read the file
    df = pd.read_csv(file)

    # Add the df to the complete df (concatenate)
    df_complete = pd.concat([df_complete, df])
    
# Reset the index
df_complete.reset_index(drop=True, inplace=True)

df_complete.drop(columns=['logits'], inplace=True)

df_complete 
```

```python
# Pivot the predictions
df_wide_pred = df_complete.pivot(
    index='subject',
    columns='task',
    values='prediction'
).reset_index()

# Get one label per subject (since they should be the same for each subject)
df_labels = df_complete[['subject', 'label']].drop_duplicates()

# Merge them together
df_wide_one = df_wide_pred.merge(df_labels, on='subject')
```

```python
# save the wide dataframe
df_wide_one.to_csv(raw_data_folder / 'alzh_pred_offline.csv', index=False)

df_wide_one
```
